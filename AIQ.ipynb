```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r'C:\Users\dell\Music\AirQuality.csv', sep=';')
df.drop(columns=['Unnamed: 15', 'Unnamed: 16'], inplace=True)
columns_to_convert = ['CO(GT)', 'C6H6(GT)', 'T', 'RH', 'AH']

for col in columns_to_convert:
    df[col] = df[col].str.replace(',', '.')
    df[col] = df[col].astype(float)

df.info()
df_cleaned = df.dropna()

print(df_cleaned.describe())



columns_to_plot = ['CO(GT)','PT08.S1(CO)','NMHC(GT)','C6H6(GT)','PT08.S2(NMHC)','NOx(GT)' ,'PT08.S3(NOx)','NO2(GT)','PT08.S4(NO2)','PT08.S5(O3)','T', 'RH', 'AH']


for col in columns_to_plot:
    plt.hist(df_cleaned[col], bins=30)
    plt.title(f'Distribution of {col}')
    plt.xlabel(f'{col} Value ------------->')
    plt.ylabel('Frequency ------------->')
    plt.grid(True)
    plt.tight_layout()
    plt.show()



```
